# Title
Silver to Gold Walkthrough
## Purpose
Show how the cleaned Silver layer becomes model-ready Gold data through the package feature-engineering helper.
## Inputs
`data/silver/listings_silver.xlsx` and the `process_silver_to_gold` helper.
## Outputs
`data/gold/listings_gold.xlsx` plus a modeling-ready `DataFrame` when the pipeline is executed.
## Key Takeaways
This notebook keeps feature engineering narrative-first and safe by default.


```python
import sys
from pathlib import Path

import pandas as pd

def resolve_project_root() -> Path:
    for candidate in [Path.cwd(), *Path.cwd().parents]:
        if (candidate / 'config.py').exists():
            return candidate
    raise FileNotFoundError('Could not find config.py in the current path ancestry')

PROJECT_ROOT = resolve_project_root()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from config import LISTINGS_GOLD, LISTINGS_SILVER
from elferspot_listings.data_processing.silver_to_gold import process_silver_to_gold

RUN_PIPELINE = False
REMOVE_PRICE_OUTLIERS = True
OUTLIER_STD = 3.0
```

```python
silver_exists = LISTINGS_SILVER.exists()
gold_exists = LISTINGS_GOLD.exists()

print(f'Silver input: {LISTINGS_SILVER}')
print(f'Silver present: {silver_exists}')
print(f'Gold target: {LISTINGS_GOLD}')
print(f'Gold present: {gold_exists}')

if silver_exists:
    silver_preview = pd.read_excel(LISTINGS_SILVER, nrows=3)
    print('\nSilver preview:')
    print(silver_preview.to_string(index=False))
else:
    print('Silver input is missing, so the walkthrough stays in dry-run mode.')
```

```python
if RUN_PIPELINE:
    gold_df = process_silver_to_gold(
        LISTINGS_SILVER,
        LISTINGS_GOLD,
        remove_price_outliers=REMOVE_PRICE_OUTLIERS,
        outlier_std=OUTLIER_STD,
    )
    print(f'Wrote {len(gold_df):,} Gold rows to {LISTINGS_GOLD}')
elif gold_exists:
    gold_df = pd.read_excel(LISTINGS_GOLD)
    print(f'Loaded existing Gold output with {len(gold_df):,} rows.')
else:
    print('Training and file writes are disabled. Enable RUN_PIPELINE only when you want to refresh Gold. ')
```
