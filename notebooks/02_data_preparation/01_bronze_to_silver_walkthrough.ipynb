# Title
Bronze to Silver Walkthrough
## Purpose
Show how the curated Bronze export is cleaned into Silver using the package transformation helper.
## Inputs
`data/bronze/listings_bronze.xlsx` and the `process_bronze_to_silver` helper.
## Outputs
`data/silver/listings_silver.xlsx` plus a cleaned `DataFrame` when the pipeline is executed.
## Key Takeaways
This notebook is a reviewer-friendly walkthrough; heavy work stays behind an explicit flag.


```python
import sys
from pathlib import Path

import pandas as pd

def resolve_project_root() -> Path:
    for candidate in [Path.cwd(), *Path.cwd().parents]:
        if (candidate / 'config.py').exists():
            return candidate
    raise FileNotFoundError('Could not find config.py in the current path ancestry')

PROJECT_ROOT = resolve_project_root()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from config import LISTINGS_BRONZE, LISTINGS_SILVER
from elferspot_listings.data_processing.bronze_to_silver import process_bronze_to_silver

RUN_PIPELINE = False
```

```python
bronze_exists = LISTINGS_BRONZE.exists()
silver_exists = LISTINGS_SILVER.exists()

print(f'Bronze file: {LISTINGS_BRONZE}')
print(f'Bronze present: {bronze_exists}')
print(f'Silver target: {LISTINGS_SILVER}')
print(f'Silver present: {silver_exists}')

if bronze_exists:
    bronze_preview = pd.read_excel(LISTINGS_BRONZE, nrows=3)
    print('\nBronze preview:')
    print(bronze_preview.to_string(index=False))
else:
    print('Bronze input is missing, so the walkthrough stays in dry-run mode.')
```

```python
if RUN_PIPELINE:
    silver_df = process_bronze_to_silver(LISTINGS_BRONZE, LISTINGS_SILVER)
    print(f'Wrote {len(silver_df):,} Silver rows to {LISTINGS_SILVER}')
elif silver_exists:
    silver_df = pd.read_excel(LISTINGS_SILVER)
    print(f'Loaded existing Silver output with {len(silver_df):,} rows.')
else:
    print('Training and file writes are disabled. Enable RUN_PIPELINE only when you want to refresh Silver. ')
```
